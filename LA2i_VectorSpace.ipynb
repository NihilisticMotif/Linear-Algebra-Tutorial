###(1) Vector Space

**Why Vector Space?**

Know the size and symetrical property of solution set.

**What is Vector Space?**

* Vector Space = set of vectors which have following property

Given that
* $V=$ set of vector
* $\vec{u},\vec{v},\vec{w}\in V$
* $r,s\in \mathbb{R}$

The addition property of $V$
* $\vec{v}+\vec{w}\in V$
* $\vec{v}+\vec{w}=\vec{w}+\vec{v}$
* $(\vec{v}+\vec{w})+\vec{u}=\vec{w}+(\vec{v}+\vec{u})$
* $\vec{v}+\vec{0}=\vec{v}$
* $\vec{v}-\vec{u}=\vec{0}$

The multiplication property of $V$
* $\vec{v}*r\in V$
* $\vec{v}*r=r+\vec{v}$
* $(\vec{v}+\vec{w})+r=\vec{w}*r+\vec{v}*r$
* $(sr)*\vec{v}=s*(r*\vec{v})$
* $1*\vec{v}=\vec{v}$
* $0*\vec{v}=\vec{0}$

**Example No.1:**

$S = \{(\frac{7}{3}-\frac{x_3}{3},\frac{-5}{3}+\frac{8x_3}{3},x_3)|x_3 \in \mathbb{R}\}$

**Addition Property**

Given that:

* $\vec{s_1}=\begin{bmatrix}
\frac{7}{3}-\frac{c_1}{3} \\
\frac{-5}{3}+\frac{8c_1}{3} \\
c_1
\end{bmatrix}
$

* $\vec{s_2}=\begin{bmatrix}
\frac{7}{3}-\frac{c_2}{3} \\
\frac{-5}{3}+\frac{8c_2}{3} \\
c_2
\end{bmatrix}
$

* $c_1,c_2,c_3\in \mathbb{R}$
* $c_1+c_2=c_3$

$\vec{s_1}+\vec{s_2}=\begin{bmatrix}
2*\frac{7}{3}-\frac{c_1+c_2}{3}\\
2*\frac{-5}{3}+\frac{8*(c_1+c_2)}{3}\\
c_1+c_2
\end{bmatrix}=\begin{bmatrix}
\frac{14}{3}-\frac{c_3}{3}\\
\frac{-10}{3}+\frac{8*(c_3)}{3}\\
c_3
\end{bmatrix}
\implies \vec{s_1}+\vec{s_2}\not\in S \implies S\text{ is not a vector space.}$

**Multiplicational Property**

Given that:

* $\vec{s_1}=\begin{bmatrix}
\frac{7}{3}-\frac{c_1}{3} \\
\frac{-5}{3}+\frac{8c_1}{3} \\
c_1
\end{bmatrix}
$
* $r,c_4\in\mathbb{R}$

$r*\vec{s_1}=\begin{bmatrix}
\frac{7r}{3}-\frac{c_1*r}{3} \\
\frac{-5r}{3}+\frac{8c_1*r}{3} \\
c_1*r
\end{bmatrix}
\neq \begin{bmatrix}
\frac{7}{3}-\frac{c_4}{3} \\
\frac{-5}{3}+\frac{8c_4}{3} \\
c_4
\end{bmatrix}\implies S\text{ is not a vector space.}$

**Example No.2:**

$S = \{(-\frac{x_3}{3},\frac{8x_3}{3},x_3)|x_3 \in \mathbb{R}\}$

**Addition Property**

Given that:

* $\vec{s_1}=\begin{bmatrix}
\frac{c_1}{3} \\
\frac{8c_1}{3} \\
c_1
\end{bmatrix}
$

* $\vec{s_2}=\begin{bmatrix}
\frac{c_2}{3} \\
\frac{8c_2}{3} \\
c_2
\end{bmatrix}
$

* $c_1,c_2,c_3\in \mathbb{R}$
* $c_1+c_2=c_3$

$\vec{s_1}+\vec{s_2}=\begin{bmatrix}
-\frac{c_1+c_2}{3}\\
\frac{8*(c_1+c_2)}{3}\\
c_1+c_2
\end{bmatrix}=\begin{bmatrix}
-\frac{c_3}{3}\\
\frac{8*(c_3)}{3}\\
c_3
\end{bmatrix}
\implies \vec{s_1}+\vec{s_2}\in S$

**Multiplicational Property**

Given that:

* $\vec{s_1}=\begin{bmatrix}
-\frac{c_1}{3} \\
\frac{8c_1}{3} \\
c_1
\end{bmatrix}
$
* $r,c_4\in\mathbb{R}$

$r*\vec{s_1}=\begin{bmatrix}
-\frac{c_1*r}{3} \\
\frac{8c_1*r}{3} \\
c_1*r
\end{bmatrix}
\neq \begin{bmatrix}
-\frac{c_4}{3} \\
\frac{8c_4}{3} \\
c_4
\end{bmatrix}\implies r*\vec{s_1}\in S \therefore S$ is a vector space.

**Example No.3:**

$S=\{a_0+a_1x+a_2x^2+...a_nx^n|a_0,...a_n\in\mathbb{R}\}$

This set is also vector space because it have all vector space property.

**Note**

It is worth noting that if solution set of augmented is
* $S=\{\} \implies S$ is not a vector space
* $S=\{\vec{p}|\vec{p}\neq\vec{0}\} \implies S$ is not a vector space
* $S=\{\vec{p}+c_1*\vec{h_1}+...+c_n*\vec{h_n}|c_1,...,c_n\in\mathbb{R},\vec{p}\neq\vec{0}\}$ is not a vector space
* $S=\{\vec{0}+c_1*\vec{h_1}+...+c_n*\vec{h_n}|c_1,...,c_n\in\mathbb{R},\}$ is a vector space
* $S=\{\vec{0}\} \implies S$ is a vector space with one solution (a.k.a. the trivial space)

###(2) Subspace

* Subspace = Vector space which are subset of vector space.

**Part of vector space**
* trivial subspace (a.k.a. $\vec{0}$)
* Itself
* Some vector space have proper subspace (a subspace of a vector space other than itself).

**Property of subspace**
* $S$ is a subspace of that vector space
* Given that $\vec{s_1},...,\vec{s_n}\in S$ and $r_1,...,r_n\in\mathbb{R} \implies \vec{s_1}*r_1+...+\vec{s_n}*r_n \in S$

###(3) Span

* Linear Span (Span) of set $S$= set of all linear combinations of the vector in $S = [S]=\text{span}(S)=\text{sp}(S)$
* span of empty set = trivial subspace

$[S]=\{\vec{s_1}*r_1+...+\vec{s_n}*r_n|r_1,...,r_n\in\mathbb{R}$ and $\vec{s_1},...,\vec{s_n}\in S\}$

**Lemma**

In a vector space, the span of any subset is a subspace.

**What is the difference between span and subspace?**

* subspace $S \in$ vector space $V \implies S \leq V \implies$ $S$ is closed under vector addition and scalar multiplication.
* $\text{sp}(S)$= span $S=\{\text{all linear combination of vector }S\}$

Relationship between Span and Subspace
* $\text{sp}(S)|S\neq\{\} \implies \text{sp}(S)$ is a subspace.
* Every subspace is $\text{sp}(A)|A$ is some vector.
* $\text{sp}(S)$ = smallest subspace $V$, $V=\{\text{all elements of }S\}$
* $\text{sp}(S)=\{\text{the intersection of all subspaces of }V\text{  that contain all the element of }S\}$
