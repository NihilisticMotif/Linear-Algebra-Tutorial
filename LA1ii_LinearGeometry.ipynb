###(1) Vector = Displacement

Keyword
* canonical representation = a standard way of presenting that object as a mathematical expression. e.g. canonical name of $\vec{v}=(2+3,5-7)$ is $\vec{v}=(5,-2)$ etc.

* Vector can be represented as displacement which composed of magnitude and a direction.
* Vectors are equal to each other when they have same magnitude and a direction (regardless of their position because they only represents displacement.). e.g. $\vec{v_1} = (4-2,5-3)$ and $\vec{v_2} = (14-12,15-13)$ are equal. etc.

Example No.1:

Go left $a$ meter, then go down $b$ meter.

$\vec{v}=$
 $\begin{pmatrix}
|a|\\
-|b|
\end{pmatrix}$

Example No.2:

In $n$D world, the UFO travels from point $(s_1,...,s_n)$ to $(f_1,...,f_n)$.

$\vec{v}=$
 $\begin{pmatrix}
f_1-s_1\\
...\\
f_n-s_n
\end{pmatrix}$

###(2) Hyperplane and Degenerate Plane in $n$ Dimensions

**Purpose**

The concept of hyperplane and degenerate plane is used for finding the space of surface in $n$ Dimensions space.

**Hyperplane**

* hyperplane is a subspace whose dimension is one less than that of its ambient space.
* Ambient space = the space surrounding an object.

If ambient space has $n$ dimensions, hyperplane has $n-1$ dimensions.

* reference: https://en.wikipedia.org/wiki/Hyperplane

**Degenerate Plane**
* degenerate case = a limiting case of a class of objects which appears to be qualitatively different from (and usually simpler than) the rest of the class, and the term degeneracy is the condition of being a degenerate case.

If ambient space has $n$ dimensions and $k < n-1$, then degenerate plane has $k$ dimensions.

* reference: https://en.wikipedia.org/wiki/Degeneracy_(mathematics)

**How to know if the point is intercepted with vector?**

Given that: 

1. Plane (which can be rather a line, hyperplane or degenerate plane) = $S = \{ \vec{p} + c_1*\vec{\beta_1} + ... + c_k*\vec{\beta_k} | c_1,...,c_k \in \mathbb{R} \}$

* $c =$ free variables
* $k =$ number of all free variables.
* $\vec{p} = $ particular solution = column vector with have only constant entries and not multiplied by any free variable.
* $(c_1*\vec{\beta_1} + ... + c_k*\vec{\beta_k}) =$ unrestricted combination

2. the target point = $\vec{a} = \begin{pmatrix}
a_1\\
...\\
a_n
\end{pmatrix}$

If there is a solution of $A$, given that:
$A=$
 $\begin{pmatrix}
p_{1,1} & \beta_{1,1} & ... & \beta_{1,k} & | & a_1\\
...\\
p_{n,1} & \beta_{n,1} & ... & \beta_{n,k} & | & a_n\\
\end{pmatrix}$ (a.k.a. if we can fine the value of every $c$ that make $S = \vec{a}$), then the target point is intercepted with the plane.
On the other hand, if there is no solution for $A$, then the target point is not intercepted with the plane.

**How to find the interception of 2 plane?:**

1. Find all value of free variable of both plane that make the value of both plane becomes equal. This can be done by Gaussian elimination. If the answer is empty set, then those 2 plane are not intercepted with each other (thus the interception of both plane does not exist.)

2. On the other hands, if the answer is not empty, then do substitute (plug-in) value of free variable in solution set of one of 2 planes.

###(4) Length, Angle and Dot Product of Vector

**Length of a Vector**

length of $\vec{v}$ = $|\vec{v}|=(v_1^2+v_2^2+...+v_n^2)^{\frac{1}{2}}=(\vec{v}\cdot\vec{v})^{\frac{1}{2}}$

Warning: $|\vec{v}|\neq |v_1|+...+|v_n|$

**Dot Product of 2 Vectors**

If $\vec{u}$ and $\vec{v}$ have the same number
of components, $\implies \vec{u}\cdot\vec{v}=u_1v_1+u_2v_2+...+u_nv_n$

If $\vec{u}$ and $\vec{v}$ do not have the same number
of components, $\implies \vec{u}\cdot\vec{v}$ = Undifined.

**Properties of Dot Product**
* $\vec{v}\cdot\vec{u}=\vec{u}\cdot\vec{v}$
* $\vec{v}\cdot(\vec{u}+\vec{w})=\vec{u}\cdot\vec{v}+\vec{u}\cdot\vec{w}$
* $\vec{0}\cdot\vec{u}=\vec{0}$
* $\vec{v}\cdot\vec{v}=(|\vec{v}|)^2$
* $\vec{v}\cdot(c*\vec{u})=c*(\vec{v}\cdot\vec{u})=\vec{u}\cdot(c*\vec{v})$

**Prove that $\vec{v}\cdot \vec{u}=\vec{u}\cdot \vec{v}$**

$\vec{v}\cdot \vec{u}=\Sigma_{i=1}^nv_iu_i=\Sigma_{i=1}^nu_iv_i=\vec{u}\cdot \vec{v}$

* $\vec{B_c}$ is column vector of $B \implies 1 \leq c \leq p$

$\therefore AB=\{\vec{A_r}\cdot \vec{B_c}\}=$
 $\begin{pmatrix}

... & ... & ... \\

\end{pmatrix}\implies AB\in\mathbb{R}^{m\times p}$

* Inaddition, vector is also a matrix.

Given that

**How to find the angle between 2 vectors?:**

* $\vec{v}$ = First Vector, $\vec{v}\neq \vec{0}$
* $|\vec{v}|$ = Length of First Vector
* $\vec{u}$ = Second Vector, $\vec{u}\neq \vec{0}$
* $|\vec{u}|$ = Length of Second Vector
* $|\vec{u}-\vec{v}|=|\Delta \vec{w}|$ = Distance between 2 Vectors

According to the law of Cosine: $c^2=a^2+b^2-2ab\cos(C)$

$\therefore |\Delta \vec{w}|^2=|\vec{v}|^2+|\vec{u}|^2-2|\vec{v}||\vec{u}|\cos(\theta)$

$2|\vec{v}||\vec{u}|\cos(\theta)=|\vec{v}|^2+|\vec{u}|^2-|\Delta \vec{w}|^2$

$\cos(\theta)=\frac{|\vec{v}|^2+|\vec{u}|^2-|\Delta \vec{w}|^2}{2|\vec{v}||\vec{u}|}$

Given that: $\cos(x)=y \implies x=\arccos(y)$

$\therefore \theta=\arccos(\frac{|\vec{v}|^2+|\vec{u}|^2-|\Delta \vec{w}|^2}{2|\vec{v}||\vec{u}|})$

$\therefore \theta=\arccos(\frac{\vec{v}\cdot \vec{u}}{|\vec{v}||\vec{u}|})$

* **In 3D case:**

$\theta=\arccos(\frac{|\vec{v}|^2+|\vec{u}|^2-|\Delta \vec{w}|^2}{2|\vec{v}||\vec{u}|})$
